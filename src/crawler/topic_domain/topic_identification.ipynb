## Actual vs Generated Topics

```python
import json
from bs4 import BeautifulSoup
with open('../dataset_domain/complete_headers.json', 'r') as openfile:
    json_object = json.load(openfile)
```

```python
states = ['california', 'kansas', 'maryland', 'massachusetts', 'new_york', 'south_carolina', 'south_dakota', 'texas', 'utah', 'washington']
```

```python
state_dict = {}
for state in states:
    state_headers = []
    child_state_headers = []
    headers = json_object[state]['headers']
    child_headers = json_object[state]['child_headers']
    # print(state, len(headers), len(child_headers))
    for header in headers:
        for value in header.values():
            # print(value, type(value))
            total_titles = value.strip('][').split(', ')
            for title in total_titles:
                soup = BeautifulSoup(title, "html.parser")
                # print(soup.text)
                state_headers.append(soup.text.strip())
    # state_dict[state] = {'headers': state_headers}
    for child_header in child_headers:
        for child_value in child_header.values():
            # print(child_value, type(child_value))
            total_child_titles = child_value.strip('][').split(', ')
            for child_title in total_child_titles:
                child_soup = BeautifulSoup(child_title, "html.parser")
                # print(child_soup.text, type(child_soup.text))
                child_state_headers.append(child_soup.text.strip())
    state_dict[state] = {'headers': state_headers, 'child_headers': child_state_headers}
print(state_dict['california']['headers'])
print(state_dict['california']['child_headers'])
```

```python
len(topic_model.get_topic_info()['zephyr']), len(topic_model.get_topic_info()['Topic'])
response_topic_dict = {}
for i in range(0, len(topic_model.get_topic_info()['zephyr'])):
    zephyr = topic_model.get_topic_info()['zephyr'][i][0].strip()
    topic = topic_model.get_topic_info()['Topic'][i]
    response_topic_dict[topic] = zephyr
```

```python
def calculate_score(topic_list):
    for topic in topic_list:
        similar_topics, similarity = topic_model.find_topics(topic, top_n=1)
        print(topic, similar_topics[0], similarity[0], response_topic_dict[similar_topics[0]])
        break
```

```python
for state in states:
    headers = state_dict[state]['headers']
    # child_headers = state_dict[state]['child_headers']
    # print(state, len(headers), len(child_headers))
    calculate_score(headers)
```

```python
for state in states:
    # headers = state_dict[state]['headers']
    child_headers = state_dict[state]['child_headers']
    # print(state, len(headers), len(child_headers))
    calculate_score(child_headers)
```

## Scoring

```python
def calculate_score(topic_list, threshold):
    average = 0
    count = 0
    for topic in topic_list:
        similar_topics, similarity = topic_model.find_topics(topic, top_n=1)
        average = average + similarity[0]
        if similarity[0] >= threshold:
            count = count + 1
    return count, average/len(topic_list)
```

```python
for state in states:
    headers = state_dict[state]['headers']
    count, average = calculate_score(headers, 0.4)
    print("Greater than threshold - {}/{} and Average is around - {}".format(count, len(headers), average))
```

```python
for state in states:
    child_headers = state_dict[state]['child_headers']
    count, average = calculate_score(child_headers, 0.3)
    print("Greater than threshold - {}/{} and Average is around - {}".format(count, len(child_headers), average))
```

```python
result = d.groupby(['states','topics']).size().reset_index().rename(columns={0: 'count'})
result['llm_response'] = "response"
result.head(10), result.columns
```

```python
for i in range(0, len(result)):
    result['llm_response'][i] = response_topic_dict[result['topics'][i]]
result.head(5)
```
